### MULTIMODEL TESTING

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from aequitas.plotting import Plot
```

```python
aqp = Plot()
```

```python
mixed_models = pd.read_csv('/Users/lorenh/Documents/DSaPP/test_multimodel_plotting.csv')
```

```python
mixed_models.loc[20:, 'model_id'] = 3
```

```python
mixed_models.model_id.unique()
```

```python
aqp.multimodel_comparison(disparities_table=mixed_models, x_metric='fpr', y_metric='fpr_disparity',
                              x_agg_method='mean', y_agg_method='mean')
plt.show()
```

```python
aqp.multimodel_comparison(disparities_table=mixed_models, x_metric='fpr', y_metric='fpr_disparity',
                              x_agg_method='mean', y_agg_method='mean', selected_models=[1,2])
plt.show()
```

```python
aqp.multimodel_comparison(disparities_table=mixed_models, x_metric='fpr', y_metric='fpr_disparity',
                              x_agg_method='median', y_agg_method='median', selected_models=[1,2])
plt.show()
```

```python
aqp.multimodel_comparison(disparities_table=mixed_models, x_metric='fpr', y_metric='fpr_disparity',
                              x_agg_method='max', y_agg_method='max')
plt.show()
```

```python
aqp.multimodel_comparison(disparities_table=mixed_models, x_metric='fpr', y_metric='fpr_disparity',
                          x_agg_method='min', y_agg_method='min', 
                          scatter_kws={'palette': sns.diverging_palette(225, 35, n=len(mixed_models.model_id.unique()))})
plt.show()
```
